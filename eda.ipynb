Introducción al Análisis Exploratorio de Datos (EDA).

El Análisis Exploratorio de Datos (EDA) es una etapa fundamental en la analítica de datos que permite resumir, visualizar y comprender los patrones subyacentes en los datos. Su propósito es:

    - Resumir los aspectos principales de los datos.
    - Entender la estructura y las relaciones dentro de los datos.
    - Detectar valores atípicos y anomalías.
    - Probar supuestos subyacentes.
    - Desarrollar conocimientos que pueden informar análisis posteriores.

Tipos de datos.

Datos Numéricos o Cuantitativos.

Estos datos consisten en valores que pueden ser medidos o contados. Los enfoques principales incluyen:

    - Medidas de Tendencia Central: Media, mediana.
    - Medidas de Dispersión: Desviación estándar, rango.

```python
import pandas as pd

# Leer datos
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/data_science_salaries.csv")

# Estadísticas descriptivas para variables numéricas
print(df.describe())

# Estadísticas descriptivas para variables categóricas
print(df.describe(include=["O"]))
```

Gráficos recomendados. 

Histograma: El histograma es útil para visualizar la distribución de los datos numéricos.

Datos categóricos o cualitativos.

Representan variables divididas en categorías distintas. Los enfoques principales incluyen:

    - Frecuencias y Proporciones: Contar cuántas veces ocurre cada categoría.

Gráfico de barras: El gráfico de barras es útil para visualizar la frecuencia de cada categoría.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['salary_in_usd'], bins=100)
plt.title('Distribución de salarios en USD')
plt.xlabel('Salario en USD')
plt.ylabel('Frecuencia')
plt.show()
```

```python
sns.countplot(x='experience_level', data=df)
plt.title('Distribución de niveles de experiencia')
plt.xlabel('Nivel de Experiencia')
plt.ylabel('Conteo')
plt.show()
```

Métodos comunes en EDA.

Estadísticas Descriptivas

Para variables numéricas:

    - Media, mediana, desviación estándar, valor mínimo y máximo.

Para variables categóricas:

    - Conteos o frecuencias de cada categoría.

Visualizaciones.

    - Variables numéricas continuas: Usar histogramas para la distribución, y diagramas de caja (box plots) para la dispersión.
    - Variables categóricas: Usar gráficos de barras para las frecuencias y gráficos de tarta para proporciones.

Relación entre variables.

    - Entre variables numéricas: Usar diagramas de dispersión y la matriz de correlación.
    - Entre variables categóricas: Usar gráficos de barras apiladas.
    - Entre variables numéricas y categóricas: Usar diagramas de caja (box plots) agrupados por categorías.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/data_science_salaries.csv")

# Verificar los tipos de datos
print(df.dtypes)

# Seleccionar solo columnas numéricas
df_numerical = df.select_dtypes(include=['float64', 'int64'])

# Calcular la matriz de correlación
corr_matrix = df_numerical.corr()

# Crear el heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

```python
sns.boxplot(x='experience_level', y='salary_in_usd', data=df)
plt.title('Distribución de salarios por nivel de experiencia')
plt.xlabel('Nivel de experiencia')
plt.ylabel('Salario en USD')
plt.show()
```

Puntos a destacar.

Tipos de datos:

    - Numéricos: Medibles o contables, analizados con medidas de tendencia central y dispersión.
    - Categóricos: Categorías distintas, analizados con conteos y proporciones.

Métodos Comunes:

    - Estadísticas descriptivas: Media, mediana para numéricos; conteos para categóricos.
    - Visualizaciones: Histogramas, diagramas de caja para numéricos; gráficos de barras, gráficos de pastel para categóricos.
    - Relación entre variables: Diagramas de dispersión, matrices de correlación para numéricos; gráficos de barras apiladas para categóricos.

Estas técnicas y gráficos permiten explorar y comprender datos, facilitando el desarrollo de conocimientos que guiarán el análisis más profundo y la toma de decisiones.
